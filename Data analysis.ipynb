```python
#Basics of Grouping of drive-wheels
```

```python
df['drive-wheels'].unique()
```

```python
df_group_one=df[['drive-wheels','body-style','price']]
```

```python
df_group_one.head(5)
```

```python
df_group_one=df_group_one.groupby(['drive-wheels'],as_index=False).mean()
```

```python
df_group_one
```

```python
df_gptest=df[['drive-wheels','body-style','price']]
```

```python
group_test1=df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
```

```python
group_test1
```

```python
grouped_piviot=group_test1.pivot(index='drive-wheels',columns='body-style')
```

```python
grouped_piviot
```

```python
grouped_piviot=grouped_piviot.fillna(0)
```

```python
grouped_piviot
```

```python
plt.pcolor(grouped_piviot,cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
#correction of color bar
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_piviot, cmap='RdBu')

#label names
row_labels = grouped_piviot.columns.levels[1]
col_labels = grouped_piviot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_piviot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_piviot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
